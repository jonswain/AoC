```python
import numpy as np

def parse_input(input_text: str) -> tuple[list[list[int]], list[str]]:
    lines = input_text.strip().split("\n")
    numbers = []
    operators = []
    for line in lines[:-1]:
        numbers.append([int(x) for x in line.split()])
    operators = lines[-1].split()
    numbers = list(map(list, zip(*numbers)))
    return numbers, operators

def parse_input_part_2(input_text: str) -> tuple[list[list[int]], list[str]]:
    input_arr = np.array([list(line) for line in input_text.split("\n")[:-1]])
    numbers = []
    column_numbers = []
    for col_id in range(input_arr.shape[1]):
        if input_arr[:, col_id].tolist() == [' ']*input_arr.shape[0]:
            numbers.append(column_numbers)
            column_numbers = []
        else:
            number_str = ''.join(input_arr[:, col_id].tolist()).strip()
            if number_str:
                column_numbers.append(int(number_str))
    numbers.append(column_numbers)
    operators = input_text.split("\n")[-1].split()
    return numbers, operators

def calculate_total(numbers: list, operators: list[str]) -> int:
    total = 0
    for i, operator in enumerate(operators):
        answer = 0
        if operator == '*':
            answer += np.prod(numbers[i])
        elif operator == '+':
            answer += sum(numbers[i])
        else:
            raise ValueError(f"Unknown operator: {operator}")
        total += answer
    return total

with open("inputs/day6.txt", "r") as f:
    input_text = f.read()

print("Part 1:", calculate_total(*parse_input(input_text)))
print(f"Part 2: {calculate_total(*parse_input_part_2(input_text))}")
```
