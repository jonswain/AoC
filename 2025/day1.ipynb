```python
from collections import Counter

with open('inputs/day1.txt', 'r') as f:
    input_text = f.read()
input_text = input_text.strip().split('\n')

def new_location(start, instruction):
    direction = instruction[0]
    distance = int(instruction[1:])
    
    if direction == 'L':
        new_position = start - distance
    elif direction == 'R':
        new_position = start + distance
    return new_position % 100

start = 50
stops = []
for instruction in input_text:
    start = new_location(start, instruction)
    stops.append(start)

counts = Counter(stops)
print(f"Part 1 Answer: {counts[0]}")

def all_locations(start, instruction):
    direction = instruction[0]
    distance = int(instruction[1:])
    
    if direction == 'L':
        new_positions = [ (start - i) % 100 for i in range(1, distance + 1) ]
    elif direction == 'R':
        new_positions = [ (start + i) % 100 for i in range(1, distance + 1) ]
    return new_positions

stops = []
start = 50
for instruction in input_text:
    new_positions = all_locations(start, instruction)
    stops += new_positions

counts =  Counter(stops)
print(f"Part 2 Answer: {counts[0]}")
```
