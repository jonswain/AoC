```python
with open("inputs/day5.txt") as f:
    input_text = f.read()

ranges, ids = input_text.split("\n\n")
ranges = ranges.splitlines()
ids = list(map(int, ids.splitlines()))

fresh = 0
for idx in ids:
    found = False
    for r in ranges:
        low, high = map(int, r.split("-"))
        if low <= idx <= high:
            found = True
            break
    if found:
        fresh += 1

print(f"Part 1: {fresh}")

unpacked_ranges = [tuple(map(int, one_range.split("-"))) for one_range in ranges]
unpacked_ranges.sort(key=lambda x: x[0])

merged_ranges = []
for lower_limit, upper_limit in unpacked_ranges:
    current_max = max([upper for _, upper in merged_ranges], default=0)
    if lower_limit > current_max:
        merged_ranges.append((lower_limit, upper_limit))
    else:
        if upper_limit > current_max:
            merged_ranges[-1] = (merged_ranges[-1][0], upper_limit)

fresh = 0
for one_range in merged_ranges:
    low, high = one_range
    fresh += high - low + 1

print(f"Part 2: {fresh}")
```
