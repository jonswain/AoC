```python
from itertools import pairwise
from collections import deque
```

```python
def turn_on_batteries(
    battery_banks: list[list[int]], number_of_batteries: int = 2
) -> int:
    total_joltage = 0
    for bank in battery_banks:
        bank = deque(bank)
        to_keep = []

        while bank:
            next_battery = bank.popleft()

            if len(to_keep) < number_of_batteries:
                to_keep.append(next_battery)

            else:
                for i, battery_pair in enumerate(pairwise(to_keep)):
                    if battery_pair[0] < battery_pair[1]:
                        to_keep.pop(i)
                        break
                if next_battery >= min(to_keep) and len(to_keep) == number_of_batteries:
                    to_keep.remove(min(to_keep))
                if len(to_keep) < number_of_batteries:
                    to_keep.append(next_battery)

        total_joltage += int("".join(map(str, to_keep)))
    return total_joltage


with open("inputs/day3.txt") as f:
    input_text = f.read()

battery_banks = [deque(int(c) for c in line) for line in input_text.strip().split("\n")]

print("Part 1:", turn_on_batteries(battery_banks, number_of_batteries=2))
print("Part 2:", turn_on_batteries(battery_banks, number_of_batteries=12))
```
